```python
# Calculator Functions 

def add(n1,n2):
    return n1 + n2 

def sub(n1,n2):
    return n1 - n2 

def mul(n1,n2):
    return n1 * n2 

def div(n1,n2):
    return n1 / n2 

# Callling functions within functions 

def calco(n1, n2, funco):
    return funco(n1, n2)

# Now calling from above

res1 = calco(99,99, mul)
rprint(res1)
```
